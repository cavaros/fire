```python
import matplotlib.pyplot as plt
import numpy as np
import json

points = np.array(json.load(open('data.json', 'r'))['points'])

factor = 0.0002

points2 = np.array([[p[0]+factor*(np.random.randint(len(points))), p[1]] for p in points])

plt.plot(points[:, 0], points[:, 1], 'o')
plt.plot(points2[:, 0], points2[:, 1], 'o')
plt.show()
```
